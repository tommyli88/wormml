# Cross-Camera Evaluation
Evaluate all 4 camera models together and generate summary tables.

**Cross-Camera Portability Note:** This notebook also demonstrates evaluating a model on data from a different camera (e.g., OG model on Tau images) to assess cross-camera generalization.

## Evaluate All 4 Cameras

```python
from wormml.evaluate import EvalConfig, evaluate

# Threshold values determined from individual pipeline tuning
configs = [
    EvalConfig(
        camera    = 'OG',
        model_path = '/runs/og/yolov11_maxacc_11l/weights/best.pt',
        images_dir = '/data/og/preprocessed/images/val',
        labels_dir = '/data/og/preprocessed/labels/val',
        conf_thr   = 0.50,  # From OG sweep_thresholds
        iou_thr    = 0.45,
    ),
    EvalConfig(
        camera    = 'Tau',
        model_path = '/runs/tau/yolov11_maxacc_11l/weights/best.pt',
        images_dir = '/data/tau/preprocessed/images/val',
        labels_dir = '/data/tau/preprocessed/labels/val',
        conf_thr   = 0.36,  # From Tau sweep_thresholds
        iou_thr    = 0.25,
    ),
    EvalConfig(
        camera    = 'LB',
        model_path = '/runs/lb/yolov11_maxacc_11l/weights/best.pt',
        images_dir = '/data/lb/preprocessed/images/val',
        labels_dir = '/data/lb/preprocessed/labels/val',
        conf_thr   = 0.265,  # From LB sweep_thresholds
        iou_thr    = 0.30,
    ),
    EvalConfig(
        camera    = 'UVA',
        model_path = '/runs/uva/yolov11_maxacc_11l/weights/best.pt',
        images_dir = '/data/uva/preprocessed/images/val',
        labels_dir = '/data/uva/preprocessed/labels/val',
        conf_thr   = 0.32,  # From UVA sweep_thresholds
        iou_thr    = 0.30,
    ),
]

results = evaluate(configs)
print('Evaluation complete!')
```

## Print Summary Tables

```python
from wormml.evaluate import print_summary_tables

print_summary_tables(results)
```

## Print LaTeX Tables

```python
from wormml.evaluate import print_latex_tables

print_latex_tables(results)
```

## Cross-Camera Portability Test (Optional)

```python
# Test OG model on Tau images to assess cross-camera generalization
cross_camera_configs = [
    EvalConfig(
        camera    = 'OG→Tau',
        model_path = '/runs/og/yolov11_maxacc_11l/weights/best.pt',  # OG model
        images_dir = '/data/tau/preprocessed/images/val',  # Tau images
        labels_dir = '/data/tau/preprocessed/labels/val',
        conf_thr   = 0.50,
        iou_thr    = 0.45,
    ),
    EvalConfig(
        camera    = 'Tau→OG',
        model_path = '/runs/tau/yolov11_maxacc_11l/weights/best.pt',  # Tau model
        images_dir = '/data/og/preprocessed/images/val',  # OG images
        labels_dir = '/data/og/preprocessed/labels/val',
        conf_thr   = 0.36,
        iou_thr    = 0.25,
    ),
]

cross_results = evaluate(cross_camera_configs)
print_summary_tables(cross_results)
```
